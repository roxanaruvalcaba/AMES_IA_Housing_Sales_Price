```python
# train/test split
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, random_state = 42)
```

```python
# Standard Scalar, though it's not recommended to do on encoded
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
```

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-3, 1, 100)

# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV(alphas = l_alphas)

# Fit model using best ridge alpha!
lasso_cv.fit(Z_train, y_train)
```

```python
print(lasso_cv.score(Z_train, y_train))
print(lasso_cv.score(Z_test, y_test))
```

```python
y_pred_train = lasso_cv.predict(X_train)
y_pred_test = lasso_cv.predict(X_test)
```
